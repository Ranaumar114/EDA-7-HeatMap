```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings


warnings.filterwarnings("ignore")
```

```python
movie = pd.read_csv('movies.csv')
credit = pd.read_csv('credits.csv')
```

```python
movie_credit = pd.merge(movie,credit, left_on="id" ,right_on = "movie_id")
movie_credit
```

```python
df = movie_credit[['budget' , 'revenue' , 'vote_average' , 'vote_count']]
df
```

```python
df = movie_credit[['budget' , 'revenue' , 'vote_average' , 'vote_count']]

#create CO - RELATION Matrix

correlation_matrix = df.corr()
correlation_matrix
```

```python
df = movie_credit[['budget' , 'revenue' , 'vote_average' , 'vote_count']]

#create CO - RELATION Matrix

correlation_matrix = df.corr()

# Create Heat Map

plt.figure(figsize = (10,6))
sns.heatmap(correlation_matrix)

plt.show()
```

```python
df = movie_credit[['budget' , 'revenue' , 'vote_average' , 'vote_count']]

#create CO - RELATION Matrix

correlation_matrix = df.corr()

# Create Heat Map

plt.figure(figsize = (10,6))

#annot creat number to give idea , # 'coolwarm' it gives graph aesthetic color , #'linewidth' to make clear box internal
sns.heatmap(correlation_matrix , annot = True , cmap = 'coolwarm' , linewidth = 0.5 )

plt.show()
```

```python
df = movie_credit[['budget' , 'revenue' , 'vote_average' , 'vote_count']]

#create CO - RELATION Matrix

correlation_matrix = df.corr()

# Create Heat Map

plt.figure(figsize = (10,6))

#annot creat number to give idea , # 'coolwarm' it gives graph aesthetic color , #'linewidth' to make clear box internal,
#'cbar' it gives colors slightly adjust
sns.heatmap(correlation_matrix , annot = True , cmap = 'coolwarm' , linewidth = 0.5 , cbar = True )

plt.show()
```

```python
df = movie_credit[['budget' , 'revenue' , 'vote_average' , 'vote_count']]

#create CO - RELATION Matrix

correlation_matrix = df.corr()

# Create Heat Map

plt.figure(figsize = (10,6))

#annot creat number to give idea , # 'coolwarm' it gives graph aesthetic color , #'linewidth' to make clear box internal,
#'cbar' it gives colors slightly adjust
sns.heatmap(correlation_matrix , annot = True , cmap = 'coolwarm' , linewidth = 0.5 , cbar = True )
plt.title('Correlation Heat Map' , fontsize = 16)

plt.show()
```
